```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

final_standing_table=[]
links=['https://www.goalzz.com/main.aspx?c=13055&stage=1&smonth=201608','https://www.goalzz.com/main.aspx?c=13055&stage=1&smonth=201609','https://www.goalzz.com/main.aspx?c=13055&stage=1&smonth=201610','https://www.goalzz.com/main.aspx?c=13055&stage=1&smonth=201611','https://www.goalzz.com/main.aspx?c=13055&stage=1&smonth=201612','https://www.goalzz.com/main.aspx?c=13055&stage=1&smonth=201701','https://www.goalzz.com/main.aspx?c=13055&stage=1&smonth=201702','https://www.goalzz.com/main.aspx?c=13055&stage=1&smonth=201703','https://www.goalzz.com/main.aspx?c=13055&stage=1&smonth=201704','https://www.goalzz.com/main.aspx?c=13055&stage=1&smonth=201705','https://www.goalzz.com/main.aspx?c=13055&stage=1&smonth=201706']

for j in range(0,len(links)):
    driver = webdriver.Chrome()
    driver.get(links[j])

    html = driver.page_source 
    driver.quit() 



    soup=BeautifulSoup(html)
    standing_table=soup.select('tr[data-match],td.match_date')
    date_index=[]
    for i in range(0,len(standing_table)):
        if str(standing_table[i]).startswith('<td'):
            standing_table[i]=standing_table[i].text
            date_index.append(i)

    for i in range(0,len(standing_table)):
        if i not in date_index:
            standing_table[i]=[standing_table[i].find('span').text[:5],standing_table[i].find_all('a', class_='tl')[0].text,standing_table[i].find('td', class_='sc').text,standing_table[i].find_all('a', class_='tl')[1].text,standing_table[i].find('td', class_='s').text]

    for i in range(0,len(standing_table)):
        if i in date_index:
            a=i
        if i not in date_index:
            standing_table[i].append(standing_table[a])

    for index in sorted(date_index, reverse=True):
        del standing_table[index]
    
    final_standing_table += standing_table
    
final_standing_table[:5]
```

```python
df = pd.DataFrame(final_standing_table, columns =['Time', 'Receiving team', 'Score', 'Guest team', 'Week','Date'])
df
```
